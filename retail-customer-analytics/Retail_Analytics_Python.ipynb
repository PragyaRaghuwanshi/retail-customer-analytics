# TASK 4: Bivariate Analysis

### Objective
Analyze the relationships between numerical variables, categorical variables, and the target variable (Response) to identify patterns and business insights.

Part A: Numerical Variables vs Response

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Response', y='Income', data=df)
plt.title('Income vs Campaign Response')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Response', y='MntWines', data=df)
plt.title('Wine Spending vs Campaign Response')
plt.show()
```

Part B: Categorical Variables vs Response

```python
pd.crosstab(df['Education'], df['Response'])
```

```python
pd.crosstab(df['Marital_Status'], df['Response'])
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Education', hue='Response', data=df)
plt.xticks(rotation=45)
plt.title('Education vs Response')
plt.show()
```

Part C: Numerical vs Categorical Variables

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Education', y='Income', data=df)
plt.xticks(rotation=45)
plt.title('Income by Education Level')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Marital_Status', y='Income', data=df)
plt.xticks(rotation=45)
plt.title('Income by Marital Status')
plt.show()
```

## Task 4 Observation

Bivariate analysis revealed meaningful relationships between customer demographics, spending behavior, and campaign responses. Customers with higher income and greater spending levels were more likely to respond positively to marketing campaigns. Education level and marital status also showed variations in response behavior. Correlation analysis highlighted positive relationships between income and product spending variables, providing valuable insights for targeted marketing strategies.
